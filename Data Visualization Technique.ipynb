```python
ax=sns.countplot(x=df['body-style'])
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(data=df,x=df['body-style'],hue='company')
sns.set(rc={'figure.figsize':(15,5)})
plt.xticks(rotation=90)
for bars in ax.containers:
    ax.bar_label(bars)
```
